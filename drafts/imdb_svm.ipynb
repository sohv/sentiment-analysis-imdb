```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
# downloading necessary NLTK data
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
df = pd.read_csv("/content/drive/MyDrive/imdb_dataset/imdb_dataset.csv")
```

```python
# loading first 5 elements of dataset
df.head()
```

# Text Preprocessing

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
# function to preprocess text using tokenize and lemmatizer
def preprocess_text(text):
    words = word_tokenize(text.lower())
    words = [lemmatizer.lemmatize(word) for word in words if word.isalpha()]
    words = [word for word in words if word not in stop_words]
    return ' '.join(words)
```

```python
df['review'] = df['review'].apply(preprocess_text)
```

```python
# Splitting dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(df['review'], df['sentiment'], test_size=0.2, random_state=42)
```

```python
# convert text data into TF-IDF features
tfidf_vectorizer = TfidfVectorizer(max_df=0.95, min_df=2, max_features=5000)
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

```python
# create and train SVM model
svm_model = SVC(kernel='linear', C=1.0, random_state=42)
svm_model.fit(X_train_tfidf, y_train)
```

```python
# make predictions on test set
y_pred = svm_model.predict(X_test_tfidf)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"SVM Accuracy: {accuracy}")
```
